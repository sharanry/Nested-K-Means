## Nested KMeans

```python
from NestedKMeans import NestedKMeans
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y = make_blobs(n_samples=100000, centers=10, n_features=2, random_state=0)
nkmeans = NestedKMeans(X, 10, 2, rho=1, earlyStop=True)
nkmeans.train(max_iter=25)
nkmeans.show()
```

## Scikit intended outputs

```python
import matplotlib.pyplot as plt
plt.cla()
plt.scatter(nkmeans.x[:, 0], nkmeans.x[:, 1], c=y, marker=".")
# plt.plot(self.c[:, 0], self.c[:, 1], 'x', color="black", markersize=15)
plt.draw()
```
